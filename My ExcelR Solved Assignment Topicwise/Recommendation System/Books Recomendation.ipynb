```python
import pandas as pd
from sklearn.metrics import pairwise_distances
import numpy as np
```

```python
book_data=pd.read_csv('books.csv')
book_data
```

```python
book_data['book_tittle'].nunique()
```

# Data Preparation

```python
book_data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
book_data
```

```python
book_data.rename(columns={'Book.Title':'book_tittle','Book.Rating':'book_rating','User.ID':'user_id'},inplace=True)
```

```python
books_pivot=pd.pivot_table(data=book_data,values='book_rating',index='user_id',columns='book_tittle').fillna(0)
```

```python
books_pivot
```

```python
books_pivot.head(50)
```

```python
cosine_data=pd.DataFrame(data=1-pairwise_distances(X=books_pivot,metric='cosine'))
cosine_data.index=book_data['user_id'].unique()
cosine_data.columns=book_data['user_id'].unique()
```

```python
cosine_data
```
