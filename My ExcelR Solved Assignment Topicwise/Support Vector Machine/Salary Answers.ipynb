# DATA PREPARATION

```python
salary_data_train.drop_duplicates(inplace=True)
salary_data_train
```

```python
cols=['workclass','education','maritalstatus','occupation','relationship','race','sex','native','Salary']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
salary_data_train[cols]=salary_data_train[cols].apply(LabelEncoder().fit_transform)
```

```python
salary_data_train.head(50)
```

# Feature Selection

```python
x=salary_data_train.drop('Salary',axis=1)
y=salary_data_train['Salary']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.20,random_state=123,stratify=y)
```

```python
from sklearn.feature_selection import RFE
from lightgbm import LGBMClassifier
```

```python
def svv(xtrain,xtest,ytrain,ytest):
    svv_cl=SVC()
    svv_cl.fit(xtrain,ytrain)
    ypred2=svv_cl.predict(xtest)
    print('Accuracy Score:',accuracy_score(ytest,ypred2))
    return 'Accuracy Score:',accuracy_score(ytest,ypred2)
```

```python
for i in range(1,14):
    rfe=RFE(estimator=LGBMClassifier(n_estimators=100),n_features_to_select=i)
    rfe.fit(xtrain,ytrain)
    rfe_train1=rfe.transform(xtrain)
    rfe_test1=rfe.transform(xtest)
    print('no of features-',i)
    svv(rfe_train1,rfe_test1,ytrain,ytest)
    print('==========================================')
```

# 5 FEATURE GIVE SAME ACCURACY AS COMPARE TO 13 FEATURES 

#    MODEL Evaluation

# Train Data

```python
Rfe_select=RFE(estimator=LGBMClassifier(n_estimators=100),n_features_to_select=5)
```

```python
Rfe_select.fit(xtrain,ytrain)
```

```python
rfe_train2=Rfe_select.transform(xtrain)
rfe_test2=Rfe_select.transform(xtest)
```

```python
rfe_train2.shape
```

```python
Svv_model=SVC()
```

```python
Svv_model.fit(rfe_train2,ytrain)
```

```python
ypred_train=Svv_model.predict(rfe_train2)
```

```python
print('Accuracy_score:',accuracy_score(ytrain,ypred_train))
```

# TEST DATA

```python
ytest_pred=Svv_model.predict(rfe_test2)
```

```python
print('Accuracy_score:',accuracy_score(ytest,ytest_pred))
```
